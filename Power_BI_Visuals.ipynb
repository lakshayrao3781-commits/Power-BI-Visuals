Question 1 : Create a Card Visual to show Total Sales.

Answer:
Steps:

Open Report View in Microsoft Power BI

Select Card Visual from Visualizations pane

Drag Sales field into the card

Power BI automatically calculates Total Sales using SUM

Result:

Displays overall sales amount in a single KPI card

Question 2  : Create a Donut Chart showing Total Sales by Ship Mode and which ship mode generates the
highest sales?

Answer:
Steps:

Select Donut Chart

Drag:

Ship Mode → Legend

Sales → Values

Enable Data Labels

Analysis:

The ship mode with the largest donut section generates the highest sales.

Question 3 : Create a Line Chart showing the Sales Trend over time (by Month and Year). Is there any
noticeable seasonal pattern in sales?

Answer:
Steps:

Select Line Chart

Drag:

Order Date → X-Axis

Sales → Y-Axis

Change Date Hierarchy to:

Year → Month

Analysis:

Sales peaks and drops can indicate seasonal demand.

Higher sales during festive/end-of-year months often show seasonality.

Compares sales across categories like:

Furniture

Technology

Office Supplies

Question 5 : Create a Matrix Visual to display Sales by Month and Category and identify the month with the
highest sales.

Answer:
Steps:

Select Matrix Visual

Drag:

Month(Order Date) → Rows

Category → Columns

Sales → Values

Analysis:

The month with the highest total sales will show the largest value in the matrix totals.

ASnswer:

Sales → Analyze

Country

Category

Sub-Category

Result:

Allows interactive drill-down analysis of sales.

Question 7 : Create a Scatter Chart to compare Discount vs. Profit by Sub-Category. Are there any outliers
where high discounts resulted in negative profit?

Answer:
Steps:

Select Scatter Chart

Drag:

Discount → X-Axis

Profit → Y-Axis

Sub-Category → Details

Sales → Size

Analysis:

Points below zero profit indicate losses.

High discount + negative profit points are considered outliers.

Question 8 : Use a Funnel Chart to show the number of orders by Customer Segment.

Answer:
Steps:

Select Funnel Chart

Drag:

Customer Segment → Group

Order ID → Values (Count)

Result:

Shows which customer segment contributes the most orders.

Question 9 : Use a Waterfall Chart to analyze how Profit changes by Category within the countries.

Answer:
Steps:

Select Waterfall Chart

Drag:

Country → Category

Profit → Y-Axis

Category → Breakdown

Result:

Shows increase/decrease in profit contribution across categories and countries.

Question 10 : Create a Dual-Axis Line Chart to show Total Sales and Profit Over Time (by Month).

Answer:
Steps:

Select Line and Clustered Column Chart

Drag:

Month(Order Date) → Shared Axis

Sales → Column Values

Profit → Line Values

Result:

Displays both:

Sales trend

Profit trend

Helps compare revenue vs profitability over time.
